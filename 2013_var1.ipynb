```python
import numpy as np
from scipy import integrate
import matplotlib.pyplot as plt

# Параметры

PI = np.pi
Mu = 4 * PI * 10**-7
lam = 200 * 10 ** -9
h = 10 * 10**-9 


# вычисление константа
def constant_term(Mu, lam, h, Is):
    return (Mu * Is )/( 8 * PI * lam * h)

# вычисление первого двойного интеграла
def first_doubleI(x, y):
    return (np.exp(-((x + l)/lam)) * ( x0 - x) )/ ( np.square(y0 - y) + np.square(x0 - x))
# вычисление второго двойного интеграла
def second_doubleI(x, y):
    return np.exp(-((l - x)/lam)) * ( x0 - x) / ( np.square(y0 - y) + np.square(x0 - x))

# пределы интеграла y
def bounds_y(h):
   return [-2*h, 0]

# пределы интеграла x1
def bounds_x1(l):
    return [-l, 0]

# пределы интеграла x2
def bounds_x2(l):
    return [0, l]

# 〈B⊥( j =1)〉 — усредненнное магнитное поле 
Is = 10 * 10 ** -6
const_term_bj1 = constant_term(Mu, lam, h, Is)

Ws = 30 * 10 ** -3
W0 = 10 * 10 ** -3
Wb = Ws
l = Wb/2

y0 = 10 * ( 10 ** -9) 
x0  = 15.5 * ( 10 ** -6)

Idxdy_1_bj1, err1 = integrate.nquad(first_doubleI, [bounds_x1(l), bounds_y(h)])
Idxdy_2_bj1 , err2 = integrate.nquad(second_doubleI, [bounds_x2(l), bounds_y(h)])
  
B_j_1 = const_term_bj1 * ( Idxdy_1_bj1 + Idxdy_2_bj1 )

# 〈B⊥( j =1)〉 
Bj1_ycped = B_j_1 / W0 

#print(Bj1_ycped)

# Ws = 30mkm W0 = 10mkm  Wp = 1mkm = Wa = Wa_1mkm
Wa_1mkm = 1 * 10 ** -3


# Wb_1 = Wb_1_1 ..... Wb_1_8
Wb_1 = 1 * 10 ** -3 
l = Wb_1 / 2

Is_1 = 0.4762 * 10 ** -6

const_term_b_1 = constant_term(Mu, lam, h, Is_1)

Idxdy_1_b_1, err1 = integrate.nquad(first_doubleI, [bounds_x1(l), bounds_y(h)])
Idxdy_2_b_1 , err2 = integrate.nquad(second_doubleI, [bounds_x2(l), bounds_y(h)])
  
B_1_1 = const_term_b_1 * ( Idxdy_1_b_1 + Idxdy_2_b_1 )


# Wb_1_9 = 2 mkm
Wb_1_9 = 2 * 10 ** -3 
l = Wb_1_9 / 2

Is_1_9 = 0.9524 * 10 ** -6
const_term_b_1_9 = constant_term(Mu, lam, h, Is_1_9)

Idxdy_1_b_1_9, err1 = integrate.nquad(first_doubleI, [bounds_x1(l), bounds_y(h)])
Idxdy_2_b_1_9 , err2 = integrate.nquad(second_doubleI, [bounds_x2(l), bounds_y(h)])
  
B_1_9 = const_term_b_1_9 * ( Idxdy_1_b_1_9 + Idxdy_2_b_1_9 )


# Wb_1_10 = 11 mkm
Wb_1_10 = 11 * 10 ** -3 
l = Wb_1_10 / 2

Is_1_10 =  5.2381 * 10 ** -6
const_term_b_1_10 = constant_term(Mu, lam, h, Is_1_10)

Idxdy_1_b_1_10, err1 = integrate.nquad(first_doubleI, [bounds_x1(l), bounds_y(h)])
Idxdy_2_b_1_10 , err2 = integrate.nquad(second_doubleI, [bounds_x2(l), bounds_y(h)])
  
B_1_10 = const_term_b_1_10 * ( Idxdy_1_b_1_10 + Idxdy_2_b_1_10 )

```

```python
B_1_sum = 2 * ((8 * B_1_1) + B_1_9 + B_1_10)

B_1_ycped = B_1_sum / Wa_1mkm

#print(B_1_ycped)


F_b_1 = B_1_ycped / Bj1_ycped

#print(F_b_1)

# F____max_массив

F_max = []

F_max.append(F_b_1)


# Ws = 30mkm W0 = 10mkm  Wp = 2mkm = Wa = Wa_2mkm
Wa_2mkm = 2 * 10 ** -3

# Wb_2 = Wb_2_1 ..... Wb_2_4
Wb_2 = 2 * 10 ** -3 
l = Wb_2 / 2

Is_2 = 1.3333 * 10 ** -6

const_term_b_2 = constant_term(Mu, lam, h, Is_2)

Idxdy_1_b_2, err1 = integrate.nquad(first_doubleI, [bounds_x1(l), bounds_y(h)])
Idxdy_2_b_2 , err2 = integrate.nquad(second_doubleI, [bounds_x2(l), bounds_y(h)])
  
B_2_1 = const_term_b_2 * ( Idxdy_1_b_2 + Idxdy_2_b_2 )


# Wb_2_5 = 14mkm
Wb_2_5 = 14 * 10 ** -3 
l = Wb_2_5 / 2

Is_2_5 = 4.6667 * 10 ** -6

const_term_b_2_5 = constant_term(Mu, lam, h, Is_2_5)

Idxdy_1_b_2_5, err1 = integrate.nquad(first_doubleI, [bounds_x1(l), bounds_y(h)])
Idxdy_2_b_2_5 , err2 = integrate.nquad(second_doubleI, [bounds_x2(l), bounds_y(h)])
  
B_2_5 = const_term_b_2_5 * ( Idxdy_1_b_2_5 + Idxdy_2_b_2_5 )

B_2_sum = 2 * ((4 * B_2_1) + B_2_5)

B_2_ycped = B_2_sum / Wa_2mkm


F_b_2 = B_2_ycped / Bj1_ycped

#print(F_b_2)

F_max.append(F_b_2)


# Ws = 30mkm W0 = 10mkm  Wp = 2mkm = Wa = Wa_5mkm
Wa_5mkm = 5 * 10 ** -3

# Wb_5 = Wb_5_1 .... Wb_5_2
Wb_5 = 5 * 10 ** -3 
l = Wb_5 / 2

Is_5 = 2.5 * 10 ** -6

const_term_b_5 = constant_term(Mu, lam, h, Is_5)

Idxdy_1_b_5, err1 = integrate.nquad(first_doubleI, [bounds_x1(l), bounds_y(h)])
Idxdy_2_b_5 , err2 = integrate.nquad(second_doubleI, [bounds_x2(l), bounds_y(h)])
  
B_5_1 = const_term_b_5 * ( Idxdy_1_b_5 + Idxdy_2_b_5 )


# Wb_5_3 = 10mkm
Wb_5_3 = 14 * 10 ** -3 
l = Wb_5_3 / 2

Is_5_3 = 5 * 10 ** -6

const_term_b_5_3 = constant_term(Mu, lam, h, Is_5_3)

Idxdy_1_b_5_3, err1 = integrate.nquad(first_doubleI, [bounds_x1(l), bounds_y(h)])
Idxdy_2_b_5_3 , err2 = integrate.nquad(second_doubleI, [bounds_x2(l), bounds_y(h)])
  
B_5_3 = const_term_b_5_3 * ( Idxdy_1_b_5_3 + Idxdy_2_b_5_3 )

B_5_sum = 2 * ((2 * B_5_1) + B_5_3)

B_5_ycped = B_5_sum / Wa_5mkm


F_b_5 = B_5_ycped / Bj1_ycped


F_max.append(F_b_5)



# Ws = 30mkm W0 = 10mkm  Wp = 2mkm = Wa = Wa_10mkm
Wa_10mkm = 5 * 10 ** -3

# Wb_10 = 10mkm
Wb_10 = 10 * 10 ** -3 
l = Wb_10 / 2

Is_10 = 5 * 10 ** -6

const_term_b_10 = constant_term(Mu, lam, h, Is_10)

Idxdy_1_b_10, err1 = integrate.nquad(first_doubleI, [bounds_x1(l), bounds_y(h)])
Idxdy_2_b_10 , err2 = integrate.nquad(second_doubleI, [bounds_x2(l), bounds_y(h)])
  
B_10 = const_term_b_10 * ( Idxdy_1_b_10 + Idxdy_2_b_10 )


B_10_sum = 2 * ((2 * B_10))

B_10_ycped = B_10_sum / Wa_10mkm


F_b_10 = B_10_ycped / Bj1_ycped


F_max.append(F_b_10)


#print(F_max)

Wp_values = [1, 2, 5, 10]
plt.plot(Wp_values, F_max , 'ko-')
plt.ylabel(' значения F_max')
plt.xlabel('значения Wp')
plt.title('граф зависимости F_max от Wp ')
plt.show()
```
