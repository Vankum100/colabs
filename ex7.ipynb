```python
import matplotlib.pyplot as plt
import numpy as np
from sympy import symbols, limit, solve

# Определение последовательности
def sequence(n):
    return (2 * n**3 + 2 * n**2 + 3) / (n**3 - 5 * n - 100)

# Построение графика
n_values = np.arange(1, 100, 1)
sequence_values = [sequence(n) for n in n_values]

plt.plot(n_values, sequence_values, label='Xn')
plt.xlabel('n')
plt.ylabel('Xn')
plt.title('График последовательности Xn')
plt.legend()
plt.show()

# Найти предел a = lim Xn as n turns to infinity
n = symbols('n')
expr = (2 * n**3 + 2 * n**2 + 3) / (n**3 - 5 * n - 100)
limit_value = limit(expr, n, np.inf)
print("Предел последовательности при n -> ∞:", limit_value)

# Вычислить n0(0.01), n0(0.001)
epsilon_1 = 0.01
epsilon_2 = 0.001

# Решение неравенства Xn - a < epsilon
n0_1 = solve(expr - limit_value - epsilon_1, n)
n0_2 = solve(expr - limit_value - epsilon_2, n)

print("n0(0.01):", n0_1[0])
print("n0(0.001):", n0_2[0])
```
