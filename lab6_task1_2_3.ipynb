```python
import matplotlib.pyplot as plt
import numpy as np

def analyze_sequence(sequence):
    n = len(sequence)

    # Построение графика последовательности
    plt.plot(range(1, n+1), sequence, marker='o')
    plt.xlabel('n')
    plt.ylabel('X_n')
    plt.title('График последовательности')
    plt.show()

    # Предположения о свойствах последовательности
    if all(x1 <= x2 for x1, x2 in zip(sequence, sequence[1:])):
        print("Последовательность монотонно неубывающая.")
    elif all(x1 >= x2 for x1, x2 in zip(sequence, sequence[1:])):
        print("Последовательность монотонно невозрастающая.")
    else:
        print("Последовательность не монотонна.")

    if all(x < float('inf') for x in sequence) and all(x > float('-inf') for x in sequence):
        print("Последовательность ограничена.")
    else:
        print("Последовательность не ограничена.")

def sequence_function_1(n):
    return (np.sin(n) + np.cos(n)) / (n * (n + 1))

def sequence_function_2(n):
    return (np.log10(10 * n)**2) / (np.log10(n)**2)

def sequence_function_3(n):
    return 1 + sum([1/k for k in range(2, n+1)])

# Создаем последовательность Xn для n от 1 до 100 (или любого другого диапазона)
n_values = np.arange(1, 101)
sequence_xn_1 = [sequence_function_1(n) for n in n_values]
sequence_xn_2 = [sequence_function_2(n) for n in n_values]
sequence_xn_3 = [sequence_function_3(n) for n in n_values]

# Анализ свойств последовательности 1
print("Анализ для последовательности 1:")
analyze_sequence(sequence_xn_1)

# Анализ свойств последовательности 2
print("\nАнализ для последовательности 2:")
analyze_sequence(sequence_xn_2)

# Анализ свойств последовательности 3
print("\nАнализ для последовательности 3:")
analyze_sequence(sequence_xn_3)
```
